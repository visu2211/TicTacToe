```python
def display_board():
    #displays tic tac toe board
    print('   |   |')
    print(' ' + board[0][0] + ' | ' + board[0][1] + ' | ' + board[0][2])
    print('   |   |')
    print('-----------')
    print('   |   |')
    print(' ' + board[1][0] + ' | ' + board[1][1] + ' | ' + board[1][2])
    print('   |   |')
    print('-----------')
    print('   |   |')
    print(' ' + board[2][0] + ' | ' + board[2][1] + ' | ' + board[2][2])
    print('   |   |')
```

```python
def player_input():
    #receives input of where player would like to place marker
    inp = True;
    result = ""
    while (inp):
        result = input("Choose a number between 1 & 9: ")
        if (1 <= int(result) <= 9 and not taken(result)):
            inp = False
        else:
            if (taken(result)):
                print(f"The #{result} slot has already been entered, please choose another number.")
            else:
                print("The character you inputted is incorrect; please try again.")
    return result
```

```python
def taken(result):
    #determines whether the inputted slot is already occupied with a marker or not
    if (result == "1"):
        if (board[0][0] == "X" or board[0][0] == "O"):
            return True
    elif (result == "2"):
        if (board[0][1] == "X" or board[0][1] == "O"):
            return True
    elif (result == "3"):
        if (board[0][2] == "X" or board[0][2] == "O"):
            return True
    elif (result == "4"):
        if (board[1][0] == "X" or board[1][0] == "O"):
            return True
    elif (result == "5"):
        if (board[1][1] == "X" or board[1][1] == "O"):
            return True
    elif (result == "6"):
        if (board[1][2] == "X" or board[1][2] == "O"):
            return True
    elif (result == "7"):
        if (board[2][0] == "X" or board[2][0] == "O"):
            return True
    elif (result == "8"):
        if (board[2][1] == "X" or board[2][1] == "O"):
            return True
    elif (result == "9"):
        if (board[2][2] == "X" or board[2][2] == "O"):
            return True
    return False
```

```python
def choose_marker():
    #players choose their marker
    marker = ''
    
    while not (marker == 'X' or marker == 'O'):
        marker = input('Player1: Choose X or O: ').upper()
    
    global p1
    global p2
    if marker == "X":
        p1 = "X"
        p2 = "O"
    else:
        p2 = "X"
        p1 = "O"
```

```python
def start_game():
    #starts the game and displays opening messages
    print("Welcome to Tic Tac Toe!")
    choose_marker()
    print("Player 1 will go first.")
    answer = " "
    while (answer != "Yes" and answer != "No"):
        answer = input("Are you ready to play? Enter Yes or No: ")
        if (answer != "Yes" and answer != "No"):
            print("Please enter the correct input. Try Again!")
        else:
            pass
    if (answer == "Yes"):
        play_game()
    else:
        print("Goodbye!")
```

```python
def check_full(): 
    #determines whether the board is full or not
    for row in board:
        for col in row:
            if (col != "X" and col != "O"):
                return False
    return True
```

```python
def check_win():
    #determines whether a player has won or not
    #r1 across
    if (board[0][0] == board[0][1] == board[0][2]):
        return True
    #r2 across
    elif (board[1][0] == board[1][1] == board[1][2]):
        return True
    #r3 across
    elif (board[2][0] == board[2][1] == board[2][2]):
        return True
    #col1 vertical
    elif (board[0][0] == board[1][0] == board[2][0]):
        return True
    #col2 vertical
    elif (board[0][1] == board[1][1] == board[2][1]):
        return True
    #col3 vertical
    elif (board[0][2] == board[1][2] == board[2][2]):
        return True
    #diagonal (top)left to right(bottom)
    elif (board[0][0] == board[1][1] == board[2][2]):
        return True
    #diagonal right to left
    elif (board[0][2] == board[1][1] == board[2][0]):
        return True
    else:
        return False
```

```python
def mark_board(response):
    #replaces number with player's marker
    global board
    sign = ""
    if (turn % 2 == 1):
        sign = p1
    else:
        sign = p2
    if (response == "1"):
        board[0][0] = sign
    elif (response == "2"):
        board[0][1] = sign
    elif (response == "3"):
        board[0][2] = sign
    elif (response == "4"):
        board[1][0] = sign
    elif (response == "5"):
        board[1][1] = sign
    elif (response == "6"):
        board[1][2] = sign
    elif (response == "7"):
        board[2][0] = sign
    elif (response == "8"):
        board[2][1] = sign
    elif (response == "9"):
        board[2][2] = sign
```

```python
def play_again():
    #asks players whether they would like to play again and starts game again
    result = input("Would you like to play again? Yes or No: ")
    if (result == "Yes"):
        global board
        board = [
            ["1", "2", "3"],
            ["4", "5", "6"],
            ["7", "8", "9"]
        ]
        global turn
        turn = 1
        play_game()
    else:
        pass
```

```python
def play_game():
    #executes all of the game playing methods
    while (check_full() == False):
        display_board()
        print()
        mark_board(player_input())
        global turn
        turn += 1
        if (check_win()):
            break
        else:
            pass
    display_board()
    if (check_full() and not check_win()):
        print("Tie!")
        play_again()
    else:
        print("Congratulations! You have won the game!")
        play_again()
```

```python
#creates the global variables referred throughout each method
board = [
        ["1", "2", "3"],
        ["4", "5", "6"],
        ["7", "8", "9"]
    ]
turn = 1
p1 = " "
p2 = " "
start_game()
```
